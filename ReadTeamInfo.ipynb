```python
# connect to database
# engine = create_engine('sqlite:///:memory:', echo=True)
engine = create_engine('sqlite:///:memory:')
Base = declarative_base()
```

```python
for i in df:
    print(i)
```

```python
class Article(Base):
    __tablename__ = 'articles'

    id = Column(Integer, primary_key=True, autoincrement=True)
    prize = Column(String)
    title = Column(String)
    logo = Column (String)
    province = Column(String)
    college = Column(String)
    brief_intro = Column(String)
    intro = Column(String)
    img1 = Column(String)
    img2 = Column(String)
    img3 = Column(String)
    img4 = Column(String)
    img5 = Column(String)
```

```python
Base.metadata.create_all(engine)
```

```python
# init session
Session = sessionmaker(bind=engine)
session = Session()
```

```python
session.query(Article).filter().first()
```

## Main operations

```python
# a function to download image
def download_img(url: str, name: str, location: str = None) -> (bool, str):
    import time
    path = os.path.join(os.getcwd(), location, name + '.jpg')
    if os.path.exists(path):
        return false, path
    try:
      with open(path, "wb") as f:
            f.write(request.urlopen(url).read())
            print("finish", url)
            return True, path
    except TimeoutError:
        print("timeout: ", url)
        return download_img(url, name, location)
    except:
        print("Unknown error", url)
        return False, "Unknown"
```

```python
# generate file name function
def rand_str(num):
    salt = ''.join(random.sample(string.ascii_uppercase + string.digits, num))
    return salt

rand_str(15)
```

```python
for num in range(max(df.count())):
    row = df.iloc[num]
    article = Article()
    article.title = row.loc["项目名称"]
    article.prize = row.loc["奖项"]
    article.province = row.loc["省份"]
    article.college = row.loc["学校"]
    article.brief_intro = row.loc["简介（300字）"]
    article.intro = row.loc["介绍（1500字）"]
    for img in range(6):
        url = df.iloc[num][img + 1]
        if pd.isna(url):
            continue
        success, path = download_img(url, rand_str(15), "Download")
        if not success:
            print("失败", num, img, path)
        else:
            if img == 0:
                article.logo = path
            elif img == 1:
                article.img1 = path
            elif img == 2:
                article.img2 = path
            elif img == 3:
                article.img3 = path
            elif img == 4:
                article.img4 = path
            elif img == 5:
                article.img5 = path
            else:
                raise FileNotFoundError
            print("Finish", num, img) 
    session.add(article)
    session.commit()
    print("Success!", path)
```
